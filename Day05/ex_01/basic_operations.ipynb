# Exercise 01 : Basic operations

```python
import pandas as pd
```

## create a dataframe views with two columns: datetime and user by reading feedviews.log

```python
df = pd.read_csv('../data/feed-views.log', sep=r'\t', engine='python', names=('datetime', 'user'))
```

- convert the datetime to the datetime64[ns] Dtype
- extract the year, month, day, hour, minute, and second from the values of that
column to the new columns

```python
df['datetime'] = pd.to_datetime(df.datetime)
df['year'] = pd.DatetimeIndex(df['datetime']).year
df['month'] = pd.DatetimeIndex(df['datetime']).month
df['day'] = pd.DatetimeIndex(df['datetime']).day
df['hour'] = pd.DatetimeIndex(df['datetime']).hour
df['minute'] = pd.DatetimeIndex(df['datetime']).minute
df['second'] = pd.DatetimeIndex(df['datetime']).second
df
```

## create the new column daytime

- you need to assign the particular time of day value if an hour is within a <br>
particular interval, for example, afternoon if the hour is larger than 11 and <br>
less or equal to 17
- 0.00 – 03.59 night, 04.00 – 06.59 early morning, 07.00 – 10.59 morning, 11.00
- 16.59 afternoon, 17.00 – 19.59 early evening, 20.00 – 23.59 evening
- use the method cut to solve this subtask

```python
df['daytime'] = pd.cut(df.hour, bins = [0, 4, 7, 11, 17, 20, 24], labels = ["night", 'early morning', 'morning', 'afternoon', 'early evening', 'evening'], include_lowest=True, right=False)
df
```

- assign the column user as the index

```python
df = df.set_index('user')
df
```

## calculate the number of elements in your dataframe

-  use the method count()

```python
df.count()
```

- calculate the number of elements in each time of day category using the method
value_counts()

```python
df['daytime'].value_counts()
```

## sort values in your dataframe by hour, minute, and second in ascending order <br> (simultaneously and not one by one)

```python
df.sort_values(ascending = True, axis=0, by =['hour', "minute","second"], inplace=True)
df
```

## calculate the minimum and maximum for the hours and the mode for the daytime<br>categories

- calculate the maximum of hour for the rows where the time of day is night

```python
max(df['hour'].where(df.daytime=='night'))
```

- calculate the minimum of hour for the rows where the time of day is morning

```python
min(df['hour'].where(df.daytime=='morning').dropna())
```

- In addition to this, find out who visited the page at those hours (make one<br>example from that)

```python
df.where(df.hour==20).dropna()
```

```python
df.where(df.hour==22).dropna()
```

- calculate the mode for the hour and daytime

```python
df['hour'].mode()
```

```python
df['daytime'].mode()
```

## show the 3 earliest hours in the morning and the corresponding usernames and the <br>3 latest hours and the usernames using nsmallest() and nlargest()

```python
df.where(df.daytime=='morning').nsmallest(3, 'hour')
```

```python
df.where(df.daytime=='morning').nlargest(3, 'hour')
```

## use the method describe() to get the basic statistics for the columns

```python
df['hour'].describe()
```

- to find out what the most popular interval for visiting the page is, calculate,<br>the interquartile range for the hour by extracting values from the result of the<br>describe() method and store it in the variable iqr

```python
iqr = df['hour'].describe()["75%"] - df['hour'].describe()["25%"]
iqr
```
